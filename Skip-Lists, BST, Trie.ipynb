**EX 2**

Insert the same values into a binary search tree in the same manner. Describe how the insertion is done and show at least a few intermediate steps in addition to the final tree. How skewed is the tree? Simulate search for 34. How many comparisons are performed?

 - Solution: In binary search tree, we starting with 1, and then add the 5. 5 > 1, therefore we add 5 to the right side. We are continuing with 7, and 7 > 5, then again we add 7 to the right side. Now we have 30. 30 > 7, therefore it is also added to the right side. New number is 2. Starting from 1, it's greater than 1, therefore we are looking at the right side of it. In the right side, we have 5, but 2 < 5. Therefore, we add it to the left side of 5. Our new number is 4. 4 < 5, so it will go to the left side of 5. There we have 2, and 2 < 4. So we add 4 to the right side of 2. Next we have 18, and it is greater than both 5 and 7, but 18 < 30. So we add 18 to the left side of 30. In the next steps, we have 32, 34, and 36, and as we see they are in ascending order. So we will add them to the right side of each other in the ascending order. As a result, the tree is skewed to the right.

 - Now I search for 34.
         1. Starting from 1, 1 < 34. Therefore, we are looking at the right side of 1. 
         2. Here we have 5, again 5 < 34.
         3. So we are looking at 7, and it is also less than 34.
         4. In the right side of 7, we have 30, and 30 < 34. So we are continuing with the right side. 
         5. Now we have 32, and it is also less than 34.
         6. In the last step, we have 34, and finally 34 = 34. It means we found the 34.
 As a result, we have 6 comparisons are performed.
