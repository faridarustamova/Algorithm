We are using the Suffix Trie and Suffix Tree for exact pattern matching, when we have many patterns and want to determine whether the text contains those patterns. However, Trie claims much more memory, while Tree compressing the letters and reduce the memory using.

Now, we are creating suffix array of "sinsinati". Our code will sort the all the suffixes and put their indexes into the array for sorted suffixes.

```python
dict_of_suffix = {"sinsinati": 0, "insinati": 1, "nsinati": 2, "sinati": 3, "inati": 4, "nati": 5, "ati": 6, "ti": 7, "i": 8}
dict_of_suffix = sorted(dict_of_suffix.items())

print("Our Suffix Array is : {", end = "")
for _ in dict_of_suffix:
    print(_[1], end = ", ")
print("}")
```

We are using the Suffix Array for pattering searching, finding the longest repeated substring, finding the longest common substring and finding the longest palindrome in a string. So, it is so useful data structure.

Although the Suffix Tree is faster at matching, the Suffix Array is more widely accepted since it takes up less space and is less complicated than the Suffix Tree. So, we can use Suffix Arrays instead of Suffix trees for this reason
