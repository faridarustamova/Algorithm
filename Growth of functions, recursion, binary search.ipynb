As we have all heard multiple times by now, binary search is an amazing algorithm on ordered/sorted data. It works by halving search space at every step, finding a solution in O(log(D)), where D is the size of the search space. In this exercise, we will implement a binary search algorithm for lexicographically sorted words.

First, download the sorted database of words we will be using at words.txt. The origin of words is "words is a standard file on Unix and Unix-like operating systems".

Given the words dictionary, your task is to:

Read in the words from a file into a suitable data structure (e.g. array of strings);
Implement a linear search algorithm. Given a word, return the index i of the word, if it exists in the word database. Return something different, e.g. -1 or None, if the word was not found. Do this by scanning through the whole word list starting from the beginning. And return the index i (location) when you see the word;
Implement a binary search algorithm. Given a word, return the index i (location) of the word, if it exists in the word database. Return something different, e.g. -1/None, if the word was not found;
Demonstrate that your implementations work! Try searching for existing words and try to search for words that do not exist.
Implementation tips:

Hint 1: Start looking at the string on index N//2, where N is the number of strings. Check whether the word is lexicographically larger or smaller. Depending on whether the word was smaller or larger, either continue your search in the left half or the right half.

Hint 2: Many programming languages have built-in support for lexicographical comparison of strings, e.g. you should be able to do “aaa” < “aab”.



 - Solution: While writing this code I used some resources. In order to get text from webpage I used "requests" and "bs4" libraries, and I refered to this resource https://matix.io/extract-text-from-webpage-using-beautifulsoup-and-python/ . Simultaneously, for binary search I refered to this resource https://www.javatpoint.com/binary-search-in-python .

```python
import requests
from bs4 import BeautifulSoup

link = 'https://raw.githubusercontent.com/PuuraJ/AdvancedAlgorithmics/master/words.txt'
page = requests.get(link).content
soup = BeautifulSoup(page, 'html.parser')
txt = soup.find(text=True)

word_list = txt.split("\n")

given_word = input("")

for i in range(len(word_list)):
    if word_list[i] == given_word:
        print(i)
        break
else:
    print("None")  
           

def binary_search(word_list, given_word):  
    low = 0  
    high = len(word_list) - 1  
    mid = 0  
  
    while low <= high:  
        # for get integer result   
        mid = (high + low) // 2  
  
        # Check if n is present at mid   
        if word_list[mid] < given_word:  
            low = mid + 1  
  
        # If n is greater, compare to the right of mid   
        elif word_list[mid] > given_word:  
            high = mid - 1  
  
        # If n is smaller, compared to the left of mid  
        else:  
            return mid  
  
            # element was not present in the list, return None  
    return "None"  
  
# Function call   
output = binary_search(word_list, given_word)
print(output)
```

This is a continuation from the previous exercise. Let’s devise an experiment to measure how much faster is our binary search algorithm compared to the linear one. For that you can:

```python
#importing the libraries
import requests
from bs4 import BeautifulSoup
import random
import time

#getting the link to extract the text from web page
link = 'https://raw.githubusercontent.com/PuuraJ/AdvancedAlgorithmics/master/words.txt'
page = requests.get(link).content
soup = BeautifulSoup(page, 'html.parser')
txt = soup.find(text=True)

#splitting the words to new list
word_list = txt.split("\n")

#getting new list to add random words from the list of words
random_words = [word_list[random.randint(0, len(word_list))] for x in range(20)]

#adding word which is not in the list of word
random_words.append("gozelem")

#a function for binary search
def binary_search(word_list, word):  
    low = 0  
    high = len(word_list) - 1  
    mid = 0  
    count = 0
    while low <= high:  
        # for get integer result   
        mid = (high + low) // 2  
        
        count += 1

        # Check if n is present at mid   
        if word_list[mid] < word:  
            low = mid + 1  

        # If n is greater, compare to the right of mid   
        elif word_list[mid] > word:  
            high = mid - 1  

        # If n is smaller, compared to the left of mid  
        else:  
            return count  
            
    return count  


#creating lists to collect time for plotting
linear_time_lst = []
binary_time_lst = []

#counting time spent for each word in the random words list 
for word in random_words:
    

    #starting the time for linear search
    start = time.time()
    
    #linear search
    for i in range(len(word_list)):
        if word_list[i] == word:
            break 
    
    #ending the time for linear search 
    end = time.time()
    
    #calculating the time spent for linear search
    time_spend = end - start               

    # starting the time for a binary search
    start1 = time.time()

    # Function call 
    output = binary_search(word_list, word)

    # ending the time for a binary search
    end1 = time.time()

    #calculating the time spent for binary search
    time_spend1 = end1 - start1

    #printing the time for each search
    print(time_spend)
    linear_time_lst.append(time_spend)
    
    print(time_spend1)
    binary_time_lst.append(time_spend1)
    
    #printing the iteration for each search
    if word in word_list:
        print(word_list.index(word) - 1)
    else:
        print(len(word_list))
    print(output)
    
```

Now I plotted the comparison of the time spents of the binary and linear search

```python
import matplotlib.pyplot as plt

X = [i for i in range(0, len(random_words))]
plt.figure(figsize=(8, 6), dpi=100)
plt.plot(X, linear_time_lst)
plt.plot(X, binary_time_lst)
plt.title("Comparison of the time spents of the binary and linear search")
plt.xlabel("Time spent")
plt.ylabel("The number of iterations")
plt.legend(["Linear", "Binary"])
plt.show()
```
